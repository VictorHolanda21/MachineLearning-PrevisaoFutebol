# 2. Preparação dos dados

## 2.1. Preparando base de classificação

```python
# Verificando campo com nomes dos times

dados_classificacao.time.unique()
```

```python
# Padronização do nome dos times tudo para minusculos na base classificação
dados_classificacao.time = dados_classificacao.time.str.lower()
dados_classificacao.time.unique()
```

```python
# retirando qualquer pontuação no nome dos times

dados_classificacao.time = dados_classificacao.time.replace("[^\w\s]"," ",regex=True)
dados_classificacao.time.unique()
```

```python
# retirando espaçoes em brancos duplicados
dados_classificacao.time = dados_classificacao.time.replace("\\s+"," ",regex=True)
dados_classificacao.time.unique()
```

## 2.2. Preparando base de elencos

```python
# Padronizando nome dos times base elenco
dados_elenco.TIME.unique()
```

```python
# Dividindo o texto com base no /
print(dados_elenco.TIME[45].split("/"))
print(dados_elenco.TIME[45].split("/")[6])
```

```python
# Aplicando o tratamento acima em todas as linhas
dados_elenco.TIME = dados_elenco.TIME.apply(lambda x: x.split("/")[6])

dados_elenco.TIME = dados_elenco.TIME.replace("[^\w\s]"," ",regex=True)

dados_elenco.TIME = dados_elenco.TIME.replace("\\s+"," ", regex=True)

dados_elenco.TIME.unique()
```

## 2.3. Campos numéricos

```python
# Corrigindo campos numéricos
dados_elenco.info()
```

```python
print(dados_elenco.IDADE.unique())
```

```python
pd.to_numeric(dados_elenco.IDADE.unique(), errors='coerce')
```

```python
colunas_numericas = [
                     'IDADE', 'ALT', 'POSICAO', 'JOGOS', 'SUBSTITUICAO', 'DEFESAS',
                     'GOLSOFRIDOS', 'ASSISTENCIAS', 'FALTASCOMETIDAS', 'FALTASOFRIDAS',
                     'CARTAOAMARELO', 'CARTAOVERMELHO', 'GOLS', 'FINALIZACOES', 'CHUTESGOL'
                     ]
```

```python
for c in colunas_numericas:
  print(c)
  dados_elenco[c] = dados_elenco[c].str.replace("[a-zA-Z]","")
  dados_elenco[c] = pd.to_numeric(dados_elenco[c],errors="coerce")
```

```python
dados_elenco.info()
```

```python
dados_elenco.head()
```

```python
dados_elenco_select = dados_elenco[[
                                    'NOME', 'POS', 'IDADE', 'ALT', 'POSICAO',
                                    'NACIONALIDADE', 'TIME', 'ANO'
                                    ]]
dados_elenco_select.head()
```

# 3. Cruzamento dos dados

```python
dados_elenco_merge = dados_elenco_select.merge(
                        dados_classificacao[['time','posicao','ano']],
                        left_on=['TIME', 'ANO'],
                        right_on=['time','ano'],
                        how='inner'                       
                      ).drop(columns=['time','ano']) # apagando colunas repetidas

dados_elenco_merge.head()
```

```python
# renomeando nome de colunas
dados_elenco_merge.rename(columns={
                              'POS':'posicao_jogador',
                              'ALT':'altura_jogador',
                              'POSICAO':'peso_jogador',
                              'posicao':'classificao_time'
                          }, inplace=True)
dados_elenco_merge.head()
```

```python
# padronizando rótulos das colunas

dados_elenco_merge.columns = [x.lower() for x in dados_elenco_merge.columns]
dados_elenco_merge.head()
```

```python
# Salvando base de dados final

dados_elenco_merge.to_csv("/content/drive/MyDrive/Colab Notebooks/Previsao No Futebol/dados_elencos_merge.csv", index=False)
```

# 4. Engenharia de Atributos (Feature Engineer)

```python
# Resumindo dados do elenco
# Calculando médias por ano e classificação do time

media_classificacao_time = dados_elenco_merge.groupby(
                                              ['ano','classificao_time']
                                            ).agg(
                                              {
                                               'idade':'mean',
                                               'altura_jogador':'mean',
                                               'peso_jogador':'mean'
                                               }
                                            ).reset_index()
media_classificacao_time.head()
```

```python
media_classificacao_time.to_csv("/content/drive/MyDrive/Colab Notebooks/Previsao No Futebol/media_classificacao_time.csv",index=False)
```
